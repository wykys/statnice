# Princip superpozice
Jde o metodu, která umožnuje řešit obvod z více zdroji. Metoda spočívá v tom, že se Vždy počítá jen s jedním zdrojem a výsledné proudy a napětí po aplikování všech zdrojů se sečtou. __Tuto metodu je možné použít pouze u lineárních obvodů!__

```python
import numpy as np
from numpy import pi, sin
from matplotlib import pyplot as plt
```

## Příklad

```python
t = np.linspace(0, 1, 1000)
U1 = 5 # V
U2 = 7 * sin(2*pi*5*t) # V
```

### 1. Lineární obvod buzen dvěma zdroji:

```python
R = 1 # Ω
I1 = U1 / R
I2 = U2 / R
I_super = I1 + I2
I_normal = (U1 + U2) / R
plt.subplot(211)
plt.plot(t, I_super)
plt.subplot(212)
plt.plot(t, I_normal, 'g')
plt.show()
```

### Chyba metody:

```python
plt.plot(I_super - I_normal, 'r')
plt.plot()
```

## 2. Obvoz z nelineárním prvkem $I = U^3$

```python
I1 = U1**3
I2 = U2**3
I_super = I1 + I2
I_normal = (U1 + U2)**3
plt.subplot(211)
plt.plot(t, I_super)
plt.subplot(212)
plt.plot(t, I_normal, 'g')
plt.show()
```

### Chyba metody:

```python
plt.plot(I_super - I_normal, 'r')
plt.plot()
```
